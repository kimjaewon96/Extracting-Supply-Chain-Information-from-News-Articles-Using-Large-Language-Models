```python
from datasets import load_from_disk, disable_caching, Dataset
import pandas as pd
import os
from pathlib import Path
```

```python
disable_caching()
cwd = os.getcwd()
ds = load_from_disk(Path(cwd, "../../datasets/ManualDataset").resolve())
df_train = ds["train"].to_pandas()
df_valid = ds["valid"].to_pandas()
df_test = ds["test"].to_pandas()
df_total = pd.concat([df_train, df_valid, df_test])
```

```python
df_total
```

```python
df_train = df_train.drop(columns="__index_level_0__")
unique_texts = df_train['original_text'].unique()
sampled_texts = pd.Series(unique_texts).sample(n=min(100, len(unique_texts)), random_state=42)
result_df = df_train[df_train['original_text'].isin(sampled_texts)]
```

```python
result_df
```

```python
Dataset.from_pandas(result_df).save_to_disk(str(Path(cwd, "../../datasets/ManualReducedDataset")))
```
